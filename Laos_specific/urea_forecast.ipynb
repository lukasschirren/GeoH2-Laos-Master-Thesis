## 1. Loading and Preparing the Data


```python
import pandas as pd
import numpy as np

# Load the provided data
file_path = 'Data\\fertiliser_data.xlsx'  # Update with your actual path
df_fertiliser = pd.read_excel(file_path)
df_fertiliser
```

```python

# For AG.PRD.FOOD.XD
last_value_food = df_fertiliser.loc[df_fertiliser['time'] == 2022, 'AG.PRD.FOOD.XD'].values[0]
values_food = [last_value_food * (1.025 ** (year - 2022)) for year in range(2023, 2030)]

# For AG.PRD.CROP.XD
last_value_crop = df_fertiliser.loc[df_fertiliser['time'] == 2022, 'AG.PRD.CROP.XD'].values[0]
values_crop = [last_value_crop * (1.025 ** (year - 2022)) for year in range(2023, 2030)]

# Update the existing rows in the DataFrame
for i, year in enumerate(range(2023, 2030)):
    df_fertiliser.loc[df_fertiliser['time'] == year, 'AG.PRD.FOOD.XD'] = values_food[i]
    df_fertiliser.loc[df_fertiliser['time'] == year, 'AG.PRD.CROP.XD'] = values_crop[i]

# Check the updated DataFrame
print(df_fertiliser[['time', 'AG.PRD.FOOD.XD', 'AG.PRD.CROP.XD']].tail(15))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the font sizes
font_size = 20  # General font size for titles and other text
axis_label_size = 22  # Font size for x and y axis labels
tick_label_size = 18  # Font size for the tick labels

# Renaming dictionary
rename_dict = {
    'AG.PRD.FOOD.XD': 'Agricultural Food Production Index',
    'AG.PRD.CROP.XD': 'Crop Production Index',
    'AG.LND.AGRI.K2': 'Agricultural Land (sq. km)',
    'fertiliser_quantity': 'Fertiliser Quantity (tons)',
    'GDP': 'GDP (current USD)'
}

# Rename the columns in the dataframe
df_renamed = df_fertiliser.rename(columns=rename_dict)

# Variables to plot
variables_to_plot = ['Agricultural Food Production Index', 'Crop Production Index', 
                     'Agricultural Land (sq. km)', 'GDP (current USD)']

# Set up the figure and axes
fig, axes = plt.subplots(1, len(variables_to_plot), figsize=(20, 5))

# Plot scatter plots with regression lines
for i, var in enumerate(variables_to_plot):
    # Calculate the correlation
    correlation = df_renamed['Fertiliser Quantity (tons)'].corr(df_renamed[var])
    
    # Plot with regression line
    sns.regplot(x=var, y='Fertiliser Quantity (tons)', data=df_renamed, line_kws={'color': 'red'}, ax=axes[i])
    
    # Set labels and title with correlation
    axes[i].set_xlabel(var, fontsize=axis_label_size)
    
    if i == 0:
        axes[i].set_ylabel('Fertiliser Quantity (tons)', fontsize=axis_label_size)
        axes[i].tick_params(axis='y', which='major', labelsize=tick_label_size)
    else:
        axes[i].set_ylabel('')
        axes[i].tick_params(axis='y', left=False, labelleft=False)
    
    axes[i].tick_params(axis='x', which='major', labelsize=tick_label_size)
    axes[i].set_title(f'Correlation: {correlation:.2f}', fontsize=font_size)

# Apply tight layout to fit everything well
plt.tight_layout()
plt.show()
```

## 3. Forecasting Models


```python
from sklearn.linear_model import LinearRegression

# Filter the data to include only rows from 1980 onwards
df_fertiliser_train = df_fertiliser[(df_fertiliser['time'] >= 1980) & (df_fertiliser['time'] <= 2019)].dropna(subset=['AG.PRD.FOOD.XD', 'AG.PRD.CROP.XD', 'GDP', 'fertiliser_quantity'])

# Check for NaN values to ensure all columns are filled
print(df_fertiliser_train[['AG.PRD.FOOD.XD', 'AG.PRD.CROP.XD', 'GDP', 'fertiliser_quantity']].isna().sum())
```

```python
# Extract relevant columns for modeling
X_train = df_fertiliser_train[['AG.PRD.FOOD.XD', 'AG.PRD.CROP.XD', 'GDP']].values
y_train = df_fertiliser_train['fertiliser_quantity'].values

# Fit a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Forecast using the updated agricultural data and GDP for 2023 to 2029
years_forecast = np.arange(2020, 2030)
X_forecast = df_fertiliser[df_fertiliser['time'].isin(years_forecast)][['AG.PRD.FOOD.XD', 'AG.PRD.CROP.XD', 'GDP']].values
forecast_fertiliser_quantity = model.predict(X_forecast)

# Prepare a DataFrame to display the forecast
forecast_df = pd.DataFrame({
    'Year': years_forecast,
    'Forecasted Fertiliser Quantity': forecast_fertiliser_quantity
})

print(forecast_df)
```

## 4. Visualization of Forecasts


```python
# Plot historical data
plt.figure(figsize=(10, 6))
plt.plot(df_fertiliser_train['time'], df_fertiliser_train['fertiliser_quantity'], label='Historical Data', color='blue')

# Plot forecasted data
plt.plot(forecast_df['Year'], forecast_df['Forecasted Fertiliser Quantity'], label='Forecast (2020-2029)', color='red', linestyle='--')

plt.title('Fertiliser Quantity Forecast (2020-2029)')
plt.xlabel('Year')
plt.ylabel('Fertiliser Quantity (tons)')
plt.legend()
plt.grid(True)
plt.show()
```
